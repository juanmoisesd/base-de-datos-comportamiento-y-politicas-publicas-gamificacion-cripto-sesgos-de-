2. Sesgos de anclaje en el diseño de subsidios verdes y disposición a pagar por tecnologías limpias.
3. Aversión al riesgo migratorio y decisiones de salida en países de Latinoamérica.

## 1. Setup


```python
# Install dependencies if needed
# !pip install pandas matplotlib seaborn numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['figure.dpi'] = 120
print('Libraries loaded.')
```

## 2. Load dataset


```python
# Download CSV from Zenodo and place it in the working directory.
# Example:
# df = pd.read_csv('base-datos-comportamiento-politicas-publicas.csv')

# Synthetic example data (for demonstration only)
np.random.seed(42)
years = list(range(2000, 2024))
keywords = [
    'gamificacion',
    'criptomonedas',
    'politicas_publicas',
    'migracion_la'
]

df = pd.DataFrame(
    {
        'year': years,
        **{k: np.random.normal(50, 15, len(years)) for k in keywords}
    }
)

print('Dataset: Base de datos: Comportamiento y Políticas Públicas')
print('DOI: 10.5281/zenodo.19017501')
print('Shape:', df.shape)
df.head()
```

## 3. Summary statistics


```python
df.describe().round(2)
```

## 4. Temporal trends


```python
fig, ax = plt.subplots(figsize=(10, 5))
for col in df.columns[1:]:
    ax.plot(df['year'], df[col], marker='o', markersize=3, label=col)

ax.set_title(
    'Tendencias temporales — Base de datos: Comportamiento y Políticas Públicas\n'
    'Gamificación cripto, subsidios verdes y migración (datos de ejemplo)'
)
ax.set_xlabel('Año')
ax.set_ylabel('Valor (escala simulada)')
ax.legend(fontsize=8)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
